```python
# Calculate letter grade of each student
def assign_letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

# find object in list of objects using the object ID attribute
def find_obj(iD, obj_list):
    for stud in obj_list:
        if stud.iD == iD:
            obj = stud
            break
    else:
        stud = None
    
    return obj

# base class: for all people, name, id, phone, email, location
class Person:
    def __init__(self, name, phone, iD, email, location):
        self.name = name
        self.iD = iD
        self.phone = phone
        self.email = email
        self.location = location


# child class: students with extra attribute: tuition fees, grade, year
class Student(Person):
    def __init__(self, name, phone, iD, email, location, fees, subs_names, subs_scores, year):
        super().__init__(name, phone, iD, email, location)
        self.fees = fees
        self.year = year
        self.subs_grades = [assign_letter_grade(score) for score in subs_scores]
        self.subs = dict(sorted(dict(zip(subs_names, zip(subs_scores, self.subs_grades))).items()))
        self.result = "Passed" if (sum(subs_scores)/(len(subs_scores)*100)) > 0.5 else "Failed"

    def __repr__(self):
        return f"name: {self.name}\nID: {self.iD}\nemail: {self.email}\nphone no.: {self.phone}\nlive in: {self.location}\nhis tuition fees: {self.fees}\nwith subjects: {self.subs}\nto the class of: {self.year}\nResult: {self.result}"

    def __str__(self):
        return f"name: {self.name}\nID: {self.iD}\nemail: {self.email}\nphone no.: {self.phone}\nlive in: {self.location}\nhis tuition fees: {self.fees}\nwith subjects: {self.subs}\nto the class of: {self.year}\nResult: {self.result}"


# child class: Teacher, with attribute: Subjects to teach, salary
class Teacher(Person):
    def __init__(self, name, phone, iD, email, location, subs, salary):
        super().__init__(name, phone, iD, email, location)
        self.subs = subs
        self.salary = salary

    def __repr__(self):
        return f"name: {self.name}\nID: {self.iD}\nemail: {self.email}\nphone no.: {self.phone}\nlive in:{self.location}\nhis subjects to teach: {self.subs}\nsalary: {self.salary}"

    def __str__(self):
        return f"name: {self.name}\nID: {self.iD}\nemail: {self.email}\nphone no.: {self.phone}\nlive in: {self.location}\nhis subjects to teach: {self.subs}\nsalary: {self.salary}"
```
